```python
# p4

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# خواندن فایل
df = pd.read_csv('dataset/Social_Network_Ads.csv')
# تبدیل ستون جنسیت به دو ستون مجزا
df = pd.get_dummies(df, columns=['Gender'], drop_first=True)
# محاسبه ماتریس همبستگی
corr_matrix = df.corr()

# بخش فوقانی ماتریس همبستگی
upper_tri = np.triu(corr_matrix)

# رسم heatmap
sns.heatmap(corr_matrix, annot=True, mask=upper_tri, cmap='coolwarm')

# تنظیمات نمودار
plt.title('Correlation between Fields')
plt.show()
```

```python
# p5
"""
pairplot یک تابع در کتابخانه seaborn است که برای رسم نمودارهای پراکندگی بین دو متغیر یا بیشتر در یک دیتافریم استفاده می‌شود. این تابع برای هر زوج از ستون‌های عددی در دیتافریم، یک نمودار پراکندگی رسم می‌کند که در قطر اصلی دیتافریم نمودارهای توزیع هر ستون به صورت هیستوگرام نمایش داده می‌شوند. همچنین با استفاده از آرگومان‌های hue و style می‌توانیم نمودارها را بر اساس یک ستون دیگر رنگ‌آمیزی کنیم.
"""


import seaborn as sns

# خواندن دیتافریم iris
iris = sns.load_dataset('iris')

# رسم نمودارهای پراکندگی بین ویژگی‌های iris
sns.pairplot(iris, hue='species')
```

```python
# p6


import seaborn as sns

# خواندن دیتافریم Social_Network_Ads
sns.set_theme(style='whitegrid')
social_network_ads = sns.load_dataset('penguins')

# رسم نمودار پراکندگی بین سن و درآمد تخمینی با توزیع چگالی احتمال (KDE)
sns.jointplot(x='bill_length_mm', y='body_mass_g', data=social_network_ads, kind='kde')

sns.jointplot(x='bill_length_mm', y='body_mass_g', data=social_network_ads, kind='hist')
```

```python
# p7

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# خواندن داده‌ها
data = pd.read_csv('dataset/Mall_Customers.csv')

# رسم هیستوگرام برای ستون 'Age'
sns.histplot(data['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# رسم هیت‌مپ برای رابطه بین 'Age' و 'Annual Income (k$)'
sns.heatmap(data[['Age', 'Annual Income (k$)']].corr(), annot=True)
plt.title('Correlation between Age and Annual Income')
plt.show()
```

```python
# p10

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
# خواندن داده‌ها
data = pd.read_csv('dataset/Mall_Customers.csv')
data['Gender'].replace({'Male': 0, 'Female': 1}, inplace=True)
# انجام PCA
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(data)

# تبدیل داده‌های PCA به دیتافریم
principalDf = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2', 'PC3'])

# رسم پلات scatter سه بعدی برای داده‌های PCA
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(principalDf['PC1'], principalDf['PC2'], principalDf['PC3'])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
plt.show()
```
